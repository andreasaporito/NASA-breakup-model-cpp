```python
import pandas as pd
import matplotlib.pyplot as plt

min_lengths = [0.001, 0.003, 0.005, 0.01, 0.02, 0.03, 0.05, 0.1, 0.2]
average_lengths = []

for ml in min_lengths:
    filename = f"../data_analysis/input-output/analysis_minlen_{ml}_result.csv"
    
    try:
        # Load the CSV file
        df = pd.read_csv(filename)
        
        # Calculate the mean of the 'Characteristic Length [m]' column
        mean_val = df['Characteristic Length [m]'].mean()
        average_lengths.append(mean_val)
        
    except FileNotFoundError:
        print(f"Warning: {filename} not found. Skipping.")
        average_lengths.append(None)
plot_data = [(m, a) for m, a in zip(min_lengths, average_lengths) if a is not None]
x_vals, y_vals = zip(*plot_data)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, marker='o', linestyle='-', color='b', linewidth=2)

plt.title(r'Average Characteristic Length vs. Minimum Threshold', fontsize=14)
plt.xlabel(r'Minimum Characteristic Length $[m]$', fontsize=12)
plt.ylabel(r'Average Fragment Characteristic Length $[m]$', fontsize=12)
plt.grid(True, which="both", ls="-", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

min_lengths = [0.001, 0.003, 0.005, 0.01, 0.02, 0.03, 0.05, 0.1, 0.2]
average_lengths = []

for ml in min_lengths:
    filename = f"../data_analysis/input-output/analysis_minlen_{ml}_result.csv"
    
    try:
        df = pd.read_csv(filename)
        mean_val = df['Characteristic Length [m]'].mean()
        average_lengths.append(mean_val)
        
    except FileNotFoundError:
        print(f"Warning: {filename} not found. Skipping.")
        average_lengths.append(None)

plot_data = [(m, a) for m, a in zip(min_lengths, average_lengths) if a is not None]
x_vals, y_vals = zip(*plot_data)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, marker='o', linestyle='-', color='b', linewidth=2)

plt.title(r'Average Characteristic Length vs. Minimum Threshold (loglog)', fontsize=14)
plt.xlabel(r'Minimum Characteristic Length $[m]$', fontsize=12)
plt.ylabel(r'Average Fragment Characteristic Length $[m]$', fontsize=12)
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.xscale('log')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
